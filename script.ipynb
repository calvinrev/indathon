jumlah_tenaga_kerja

```python
jumlah_tenaga_kerja = pd.read_csv('jumlah_tenaga_kerja_dki.csv')

# Fill the missing value with EMWA (Exponentially Weighted Moving Average)
jumlah_tenaga_kerja['pekerja_formal_ewma'] = jumlah_tenaga_kerja[['pekerja_formal']].apply(lambda x: x.ewm(span=2).mean()).astype('int64')
jumlah_tenaga_kerja['pekerja_informal_ewma'] = jumlah_tenaga_kerja[['pekerja_informal']].apply(lambda x: x.ewm(span=2).mean()).astype('int64')
jumlah_tenaga_kerja['penduduk_bekerja_ewma'] = jumlah_tenaga_kerja[['penduduk_bekerja']].apply(lambda x: x.ewm(span=2).mean()).astype('int64')
jumlah_tenaga_kerja.head(12)
```

JOIN & PROCESS DATA

```python
df = training_jumlah_penumpang.copy()
df = df.merge(jumlah_hari_libur, on=['tahun','month'])
df = df.merge(jumlah_armada_tj, on=['tahun','month'])
df = df.merge(jumlah_penumpang_mrt, on=['tahun','month'])
df = df.merge(jumlah_penumpang_lrt, on=['tahun','month'])
df = df.merge(jumlah_perjalanan_mrt, on=['tahun','month'])
df = df.merge(jumlah_perjalanan_lrt, on=['tahun','month'])
df = df.merge(jumlah_tenaga_kerja[['tahun','pekerja_formal_ewma','pekerja_informal_ewma','penduduk_bekerja_ewma']], on='tahun',how='left')
del df['tahun']
print(df.shape)
df.tail()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
scaler = MinMaxScaler()

df_to_normalized = df.copy()
del df_to_normalized['month']
del df_to_normalized['jumlah_penumpang']

# Fit and transform the data
df_to_normalized = pd.DataFrame(scaler.fit_transform(df_to_normalized), columns=df_to_normalized.columns)
df_normalized = pd.concat([df[['month']], df_to_normalized,df['jumlah_penumpang']], axis=1)
df_normalized.tail()
```

```python
# correlation
df_normalized.loc[:, df_normalized.columns != 'month'].corr()
```

```python
# del df_normalized['total_hari']
del df_normalized['penduduk_bekerja_ewma']
```

MODELLING

```python
df_normalized_testing = df_normalized.iloc[-3:].copy()
df_normalized_testing['ds'] = pd.to_datetime(df_normalized_testing['month'], format='%Y-%m-%d')
df_check = df_normalized_testing[['ds','jumlah_penumpang']].copy()
del df_normalized_testing['month']
del df_normalized_testing['jumlah_penumpang']
df_normalized_testing.info()
```

```python
# Rename columns to fit Prophet's requirements
df_normalized = df_normalized.rename(columns={'month': 'ds', 'jumlah_penumpang': 'y'})

# Create model and fit to data
# model = Prophet(seasonality_mode='additive')
model = Prophet(seasonality_mode='additive', changepoint_prior_scale=0.04)

# Add Regressor
# Add Regressor
model.add_regressor('jumlah_hari_libur_cuti')
model.add_regressor('total_hari')
model.add_regressor('jumlah_armada_tj')
model.add_regressor('jumlah_penumpang_mrt')
model.add_regressor('jumlah_penumpang_lrt')
model.add_regressor('jumlah_perjalanan_mrt')
model.add_regressor('jumlah_perjalanan_lrt')
model.add_regressor('pekerja_formal_ewma')
model.add_regressor('pekerja_informal_ewma')
```

```python
# Model
model.fit(df_normalized)
```

```python
 # Create future dataframe with desired forecasting period
future = model.make_future_dataframe(periods=3, freq='MS')
future = future.merge(df_normalized_testing, on = 'ds', how = 'left')
future = future.dropna().reset_index(drop=True)
future
```

```python
# Make predictions
forecast = model.predict(future)
forecast = forecast[['ds','yhat']]
forecast['yhat'] = forecast['yhat'].astype('int64')
forecast
```

```python
result = forecast.merge(df_check, on='ds')
print(math.sqrt(mean_squared_error(result['yhat'], result['jumlah_penumpang'])))
result.head(6)
```
